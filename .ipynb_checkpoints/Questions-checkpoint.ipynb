## 3. Первый и второй законы Ньютона. Инерциальные системы отсчета. Масса как мера инерции. Закон сохранения массы (Вывод).

Динамика область физики, которая занимается изучением того как тела влияют друг на друга, и законами их взаимодействия. Законы динамики были открыты Ньютоном и легли в основу классической механики. классическая механика действет при скоростях $v$, таких что $v << c$

Рассмотрим тело на которое не действуют никакие другие тела. Т.к. движение относительно в разных системах отчета это тело может двигаться по разному: покоиться, двигаться равномерно без ускорения или ускорено. 

Первый закон Ньютона выделяет такие системы отчета в которых изолированное тело сохраняет свою скорость по направлению и модулю (По сути своей повторяет принцип инертности Галилея).  Такие системы называются инерциальными. Инерция - свойство тел сохранять сою скорость, при отсутствии действия на них других тел.

Инерциальными являются например:
* система отсчета связанная с Землей, приближенно инерциальна, для рассчетов движения тел вблизи Земли, но например маятнику Фуко берет и все обламывает.
* гелиоцентрическая система отсчета (центр - центр солнца), более точная использовалась при формулировке закона всемирного тяготения.
* другие ситемы с разной точностью можно считать инерциальными

Системы из класса инерциальных систем движутся относительно друг друга равномерно, прямолинейно.

Итак в инерциальных системах тело меняет свою скорость при взаимодействии с другими телами. Т.е. впроцессе взаимодействия тела приобретают ускорения. Мерой инертности тела является масса (В СИ измеряется в кг).
$$\frac{m_1}{m_2} = -\frac{a_2}{a_1}$$
Т.е. ускорения полученные при взаимодействии обратнопропорциональны массам. Чтобы узнать массу конкретного тела можно сравнить его с эталоном ($m_{эт} = 1 кг$): 
$$\frac{m_1}{m_{эт}} = -\frac{a_{эт}}{a_1}$$
$$m_1 = -\frac{a_{эт}}{a_1}кг$$ 

Масса аддитивная величина, т.е. в закрытой физической системе сумма масс компонентов - $const$

Сила - мера взаимодействия тел/причина изменения скорости тела, измеряется в Ньютонах. Сила величиной в 1Н сообщает телу массой 1кг ускорение в 1$\frac{м}{с^2}$

Второй закон Ньютона обобщает следующие наблюдения:
1. Если $F - const$, то $a \propto \frac{1}{m}$
2. Если $m - const$, то $\vec{F} \propto \vec{a}$

$$\vec{F} = m\vec{a}$$
где $\vec{F}$ - равнодействующая сил, которая является суммой всех сил действующих на тело.

## 4. Третий закон Ньютона. Законы сохранения импульса и момента импульса.

По одному из свойств массы:
$$\frac{m_1}{m_2}=-\frac{a_2}{a_1}$$
Второй закон Ньютона:
$$\vec{F_1} = m_1\vec{a_1}$$
$$\vec{F_2} = m_2\vec{a_2}$$
Из этого получаем третий закон Ньютона:
$$\vec{F_1} = - \vec{F_2}$$
Тела действуют друг на друга с силами равными по значению и противоположными по направлению.

Импульс - произведение массы тела на скорость тела.
Закон сохранения импульса следует из третьего закона Ньютона и того факта, что произведение импульса по времени - равно силе (сила характеризует скорость изменения импульса)
Момент импульса - векторное произведение радиусвектора до тела и импульса тела
Момент имульса системы - векторная сумма моментов импульса для элементов системы. В изолированной инерциальной системе - момент импульса $const$

## 5. Работа, кинетическая и потенциальная энергия. Сохранение механической энергии.

Работа - мера действия силы, на тело.
Вычисляется по следующей формуле:
$$A = F  s  \cos (F, s) = \vec{F} \cdot \vec{s}$$
Для изменяющихся сил и движения:
$$A = \int \vec{F} \vec{ds}$$

Энергия – это физическая величина, характеризующая способность тела совершать работу. Есть много разных видов энергии, в динамике используются в основном кинетическая и потенциальная.

Кинетическая - энергия движущегося тела. Искать через второй закон Ньютона выразив разницу между тем, что было и тем, что стало.

Потенциальная энергия - общее название для энергии связанной с расположением чего-то по отношению к чему-то, Например потенциальная энергия тяготения в поле Земли, или потенциальная энергия сжатой пружины.

Сумма энергии кинетической и потенциальной называется механической энергией системы. Для изолированной системы это значение не изменяется. Если же энергия меняется она либо тратится на совершение работы, либо с системой совершают работу. К слову из-за этого самого закона сохранения не видать на вечного двигателя.

## 6. Абсолютно упругий и абсолютно неупругий удар. Применение законов сохранения.

Удар - столкновение двух тел, когда время взаимодействия очень мало. 
Абсолютно упругий удар - два тела сталккиваются и их общая кинетическая энергия остается прежней.
Абсолютно неупругий удар - удар при котором тела становятся единым целым, при этом теряется энергия (найти можно так же через формулы кинетической энергии до и после).

## 8. Гармонические колебания. Вынужденные и затухающие колебания. Математический маятник.

Гармонические колебания - колебания, при которых физическая величина изменяется по гармоническому (sin или cos) закону.
Например:
$$x(t) = A \sin (\omega t + \phi)$$
$A$ - амплитуда колебания  
$\omega$ - циклическая частота, которая показывает на сколько радиан смещается за секунду фаза колебания  
$\phi$ - начальная фаза  
Так же можно описать дифф.уром:
$$\frac{d^2 x}{d t^2} + \omega^2 x = 0$$
любое нетривиальное решение этого уравнения - гармоническое колебание с циклической частотой омега.

Диссипация энергии - переход части энергии системы в энергию или теплоту. Если в системе присутствует ненулевая диссипация, то колебания будут затухающими, характеризуется это в первую очередь уменьшением амплитуды.

ВЫнужденные колебания и автоколебания - и то и другое колебания вызванные воздействием внешних сил разница в том что для вынужденных внешние силы воздействуют периодически и частота вынужденных колебаний такая же что и у этих внешних колебаний. В системе с вынужденными колебаниями может возникнуть резонанс - неограниченный рост амплитуды  при совпадении частоты вынуждающей силы и частоты свободных колебаний.

Для автоколебаний - пример: маятник в дедушкиных часах, которому помогают (непереодически!) грузики.

Математический маятник - небезызвестный груз на невесомой тонкой нерастягивающейся нити, находящийся в поле тяготения. Однородном. Ему сообщают потенциальную энергию и отпускают в точке покоя он успел набрать кинетическую энергию и поэтому продолжает движение. Поднимается на свою начальную высоту (если конечно система не диссипативна) и оттуда снова начинает движение вниз.
